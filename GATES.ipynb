## NOR GATE

```python
import numpy as np

w1=int(input("Enter the Weight w1:"))
w2=int(input("Enter the Weight w2:"))
bias=int(input("Enter the bias term:"))
w=[w1,w2]
coordinates=[(x,y) for x in range(2) for y in range(2)]
print(coordinates)
norgate=[int(not(x or y)) for x in range(2) for y in range(2)]
print(norgate)

eff=0
i=0
for r in coordinates:
    z=np.dot(r,w)+bias
    if z>0:
        out=1
    else:
        out=0
    if out==norgate[i]:
        eff=eff+1
    i=i+1
    
per=(eff/4)*100
print("Efficiency:", per, "%")
```

### NAND GATE

```python
import numpy as np

w1=int(input("Enter the Weight w1: "))
w2=int(input("Enter the Weight w2: "))
bias=int(input("Enter the Bias Term:"))
w=[w1, w2]
coordinates=[(x, y) for x in range(2) for y in range(2)]
print(coordinates)

nandgate=[int(not (x and y)) for x in range(2) for y in range(2)]
print(nandgate)

eff=0
i=0
for r in coordinates:
    z=np.dot(r, w)+bias
    out=1 if z > 0 else 0
    if out == nandgate[i]:
        eff += 1
    i += 1

per = (eff / 4) * 100
print("NAND Gate Efficiency:", per, "%")
```

```python
import textwrap

def highlight_text(text, highlight=True):
    if highlight:
        return f'\033[1m{text}\033[0m'  # Bold text
    return text

def format_datasets_info():
    # Common headings
    description_heading = "Description"
    features_heading = "Features"
    classes_heading = "Classes"
    sample_heading = "Sample"
    sources_heading = "Source"

    # Raisin Dataset
    raisin_main_heading = "Raisin Dataset"
    raisin_description = "The Raisin Dataset contains features extracted from images of raisin varieties to classify them based on their physical characteristics."
    raisin_features = [
        "Area",
        "Perimeter",
        "Major Axis Length",
        "Minor Axis Length",
        "Eccentricity",
        "Convex Area",
        "Extent",
        "Solidity",
        "Equivalent Diameter",
        "Orientation"
    ]
    raisin_classes = [
        "Kecimen (Class 0)",
        "Besni (Class 1)"
    ]
    raisin_sample = "900"
    raisin_sources = "https://archive.ics.uci.edu/dataset/850/raisin"
    
    # Mushroom Dataset
    mushroom_main_heading = "Mushroom Dataset"
    mushroom_description = "Contains characteristics of various mushroom species, including whether they are edible or poisonous."
    mushroom_features = [
        "Cap Shape",
        "Cap Color",
        "Gill Color",
        "Gill Size",
        "Gill Attachment",
        "Gill Spacing",
        "Stalk Shape",
        "Stalk Root",
        "Stalk Surface Above Ring",
        "Stalk Surface Below Ring",
        "Stalk Color Above Ring",
        "Stalk Color Below Ring",
        "Ring Number",
        "Ring Type",
        "Spore Print Color",
        "Population",
        "Habitat"
    ]
    mushroom_classes = [
        "Edible",
        "Poisonous"
    ]
    mushroom_sample = "8,124"
    mushroom_sources = "https://www.kaggle.com/datasets/uciml/mushroom-classification"
    
    # Wine Quality Dataset
    wine_main_heading = "Wine Quality Dataset"
    wine_description = "Contains data on chemical properties of red and white wines to predict wine quality."
    wine_features = [
        "Fixed Acidity",
        "Volatile Acidity",
        "Citric Acid",
        "Residual Sugar",
        "Chlorides",
        "Free Sulfur Dioxide",
        "Total Sulfur Dioxide",
        "Density",
        "pH",
        "Sulphates",
        "Alcohol"
    ]
    wine_classes = [
        "Quality (rating from 0 to 10)"
    ]
    wine_sample = "6,497"
    wine_sources = "https://archive.ics.uci.edu/ml/datasets/Wine+Quality"
    
    # Cardiovascular Disease Dataset
    cardio_main_heading = "Cardiovascular Disease Dataset"
    cardio_description = "Contains health-related information used to predict the likelihood of cardiovascular disease in patients."
    cardio_features = [
        "Age",
        "Gender",
        "Blood Pressure",
        "Cholesterol Levels",
        "Glucose Levels",
        "Other Risk Factors"
    ]
    cardio_classes = [
        "Presence or absence of cardiovascular disease (binary classification: 1 for presence, 0 for absence)"
    ]
    cardio_sample = "Approximately 70,000"
    cardio_sources = "https://www.kaggle.com/datasets/sulianova/cardiovascular-disease-dataset"
    
    # Bank Churn Prediction Dataset
    bank_main_heading = "Bank Churn Prediction Dataset"
    bank_description = "Contains data on bank customers to predict whether they will churn (i.e., stop using the bank's services) based on their account activities and demographic information."
    bank_features = [
        "Customer ID",
        "Account Balance",
        "Number of Transactions",
        "Age",
        "Tenure",
        "Product Usage",
        "Other Features Relevant to Customer Behavior"
    ]
    bank_classes = [
        "Churn status (binary classification: 1 for churned, 0 for not churned)"
    ]
    bank_sample = "Typically around 10,000 to 100,000"
    bank_sources = [
        "https://www.kaggle.com/datasets",
    ]
    
    # Formatting Raisin Dataset
    print(highlight_text(raisin_main_heading))
    print()
    print(f'{highlight_text(description_heading, False)}: {raisin_description}')
    print()
    print(f'{highlight_text(features_heading, False)}:')
    for feature in raisin_features:
        print(f' . {feature}')
    print()
    print(f'{highlight_text(classes_heading, False)}:')
    for cls in raisin_classes:
        print(f' . {cls}')
    print()
    print(f'{highlight_text(sample_heading, False)}: {raisin_sample}')
    print()
    print(f'{highlight_text(sources_heading, False)}: {raisin_sources}')
    print()
    
    # Formatting Mushroom Dataset
    print(highlight_text(mushroom_main_heading))
    print()
    print(f'{highlight_text(description_heading, False)}: {mushroom_description}')
    print()
    print(f'{highlight_text(features_heading, False)}:')
    for feature in mushroom_features:
        print(f' . {feature}')
    print()
    print(f'{highlight_text(classes_heading, False)}:')
    for cls in mushroom_classes:
        print(f' . {cls}')
    print()
    print(f'{highlight_text(sample_heading, False)}: {mushroom_sample}')
    print()
    print(f'{highlight_text(sources_heading, False)}: {mushroom_sources}')
    print()
    
    # Formatting Wine Quality Dataset
    print(highlight_text(wine_main_heading))
    print()
    print(f'{highlight_text(description_heading, False)}: {wine_description}')
    print()
    print(f'{highlight_text(features_heading, False)}:')
    for feature in wine_features:
        print(f' . {feature}')
    print()
    print(f'{highlight_text(classes_heading, False)}:')
    for cls in wine_classes:
        print(f' . {cls}')
    print()
    print(f'{highlight_text(sample_heading, False)}: {wine_sample}')
    print()
    print(f'{highlight_text(sources_heading, False)}: {wine_sources}')
    print()
    
    # Formatting Cardiovascular Disease Dataset
    print(highlight_text(cardio_main_heading))
    print()
    print(f'{highlight_text(description_heading, False)}: {cardio_description}')
    print()
    print(f'{highlight_text(features_heading, False)}:')
    for feature in cardio_features:
        print(f' . {feature}')
    print()
    print(f'{highlight_text(classes_heading, False)}:')
    for cls in cardio_classes:
        print(f' . {cls}')
    print()
    print(f'{highlight_text(sample_heading, False)}: {cardio_sample}')
    print()
    print(f'{highlight_text(sources_heading, False)}: {cardio_sources}')
    print()
    
    # Formatting Bank Churn Prediction Dataset
    print(highlight_text(bank_main_heading))
    print()
    print(f'{highlight_text(description_heading, False)}: {bank_description}')
    print()
    print(f'{highlight_text(features_heading, False)}:')
    for feature in bank_features:
        print(f' . {feature}')
    print()
    print(f'{highlight_text(classes_heading, False)}:')
    for cls in bank_classes:
        print(f' . {cls}')
    print()
    print(f'{highlight_text(sample_heading, False)}: {bank_sample}')
    print()
    print(f'{highlight_text(sources_heading, False)}:')
    for source in bank_sources:
        print(f' . {source}')

format_datasets_info()
```
