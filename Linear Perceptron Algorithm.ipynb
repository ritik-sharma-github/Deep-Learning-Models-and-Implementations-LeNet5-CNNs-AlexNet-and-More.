```python
w1 = int(input("Enter the weight w1: "))
w2 = int(input("Enter the weight w2: "))
w3=int(input("Enter the weight w3: "))
bias = float(input("Enter the initial bias: "))
eta = float(input("Enter the learning rate (eta): "))
w = np.array([bias, w1, w2, w3])

i=1
df1 = pd.DataFrame(columns=['bias', 'w1', 'w2', 'w3'])

for x in range(0,27):
    row=[1]
    for y in range(0,3):
        row.append(df.iloc[x,y])
    output=np.dot(w,row)
    if output >= 0:
        y1=1
    else:
        y1=-1
    
    d=df.iloc[x,3]
    w=w+eta*(d-y1)*x
    print(w)
    df1.loc[i]=w
    i=i+1

    
output_filename = "updated_weights.xlsx"
df1.to_excel(output_filename, index=False)
print("Dataframe is written to Excel file successfully")
```
