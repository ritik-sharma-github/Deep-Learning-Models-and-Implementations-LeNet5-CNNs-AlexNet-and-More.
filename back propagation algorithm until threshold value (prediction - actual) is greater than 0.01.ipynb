```python
import numpy as np
import random as rn
import tensorflow as tf
import pandas as pd
```

#### With the While loop

```python
i1=float(input("Enter the first input:"))           # enter neurons as float
i2=float(input("Enter the second input: "))
actual=float(input("Enter the actual output: "))    # Enter the Actual output
rate=float(input("Enter the learning rate: "))      # Enter the Learning Rate

I=[i1,i2]                                     # make it into list
i=np.array(I)                                 #convert it into array
print(i)


l1=[rn.random() for x in range(4)]          # generate 4 random numbers for weights w1,w2,w3,w4
w1=np.array(l1)                             # converts weight into an array
w=w1.reshape(2,2)                           # convert it shape into matrix that is 2*2
print(w)


l2=[rn.random() for x in range(2)]          # generate 2 random numbers for weight for output layer
w2=np.array(l2)                             # convert it into array
print(w2)

threshold=0.01
delta=1

while abs(delta) > threshold:
    h=np.dot(i, w)                               # take dot product of the input and weights to find the hidden layers value
    ah=tf.keras.activations.tanh(h)      # Apply the activation function
    
    
    out=np.dot(w2,ah)
    o=tf.keras.activations.tanh(out)
    
    
    delta=o-actual                                           # Error Prediction value
    print("Error at the output layer is:", delta)
    
    
    # Back-Propagation Algorithm
    W2=w2-(rate*delta*ah)                                  
    print("Updated weight for output layer :", W2)
    w2=np.array(W2)
    w2=w2.reshape(1,2)
    
    temp=rate*delta*(w2.T*i)
    W1=w-temp
    print("updated weights for hidden layer:", W1)
    
    # Update weights for next iteration
    w2 = W2
    w = W1
    
    
print("Training complete.")
print("Final weights for hidden layer:", w)
print("Final weights for output layer:", w2)
```
